Types of arguments
1. positional args
2. arbitary args
3. defualt args 

```python
# *args
def greeting(* args):
    for name in args:
        print(f'hello, {name}')

greeting('shekhar')
greeting('shehar','jatin','ik')
```

```python
# **kwargs
def printDetails(** kwargs):
    for key,value in kwargs.items():
        print(f'{key} : {value}')

printDetails(name='Alice',age=20)
printDetails(name="Bob", age=25, location="New York")
```

```python
def sum(* num):
    s=0
    for i in num:
        s = s + i
        
    return s    


sum(1,2,3,4)    
```

```python
 # print only salary if present
    # else print "No salary"
def employees(**kwargs):
    if 'Salary' in kwargs:
        print(kwargs['Name'])
    else:
        print(f'No Salary for {kwargs['Name']}')    


employees(Name='Shekhar',Salary=20000)
employees(Name='Jatin')
```

```python
def employees_2(**kwargs):
    salary = kwargs.get('Salary')
    if salary is not None:
        print(salary)
    else:
        print(f'No salary for {kwargs['Name']}')    

employees_2(Name='Shekhar',Salary=30000)     
employees_2(Name='Jatin')   
```

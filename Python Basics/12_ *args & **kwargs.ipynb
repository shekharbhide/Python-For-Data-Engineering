Types of arguments
1. positional args
2. arbitary args
3. defualt args 

```python
# *args

def greeting(* args):
    for name in args:
        print(f'hello, {name}')

greeting('shekhar')
greeting('shehar','jatin','ik')
```

```python
# **kwargs

def printDetails(** kwargs):
    for key,value in kwargs.items():
        print(f'{key} : {value}')

printDetails(name='Alice',age=20)
printDetails(name="Bob", age=25, location="New York")



```
